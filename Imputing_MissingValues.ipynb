```python
# Step 1: Import Required Libraries
import pandas as pd
```

# Load the Dataset Into a Dataframe

```python
# Step 2: Load the Dataset Into a Dataframe
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

```python
# Step 3: Define the structure of dataframe
df.shape
```

# Finding and Removing Duplicates

```python
# *** Identify duplicate rows in the dataset ****

# Step 1 — Check whether any duplicate rows exist
df.duplicated().any()
```

```python
# Step 2 — Count total duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
# Step 3 — View the duplicate rows (if you want to inspect them)
df[df.duplicated()]
```

```python
# Step 4: Remove duplicate rows

df.drop_duplicates(inplace=True)
print("Duplicate rows removed successfully.")

# Verify the change
print("New shape of dataset:", df.shape)
```

# Step 4: Finding Missing Values

```python

# Step 1: Count missing values column-wise
missing_values = df.isnull().sum()
print(missing_values)

```

```python
# Step 2: Show only columns that contain missing values (cleaner view)
missing_columns = df.isnull().sum()[df.isnull().sum() > 0]
print(missing_columns)
```

```python
# Step 3:  Count total missing values in the dataset
total_missing =df.isnull().sum().sum()
print("Total missing values in dataset:", total_missing)
```

```python
# Step 4: Count missing values in RemoteWork
missing_remote = df["RemoteWork"].isnull().sum()
print("Number of missing rows in RemoteWork:", missing_remote)
```

# Step 5. Imputing Missing Values

```python
# Step 1:  Find the value counts for the column RemoteWork
majority_remote = df["RemoteWork"].mode()[0]
print("Most frequent value in RemoteWork:", majority_remote)
```

```python
# Step 2: Impute missing values in RemoteWork using the majority value
df["RemoteWork"].fillna(majority_remote, inplace=True)
print("Missing values in RemoteWork have been replaced with the majority value.")
```

## --------Identify compensation-related columns--------


```python
# Step 3:  Check for any compensation-related columns and describe their distribution

comp_cols = [col for col in df.columns if "Comp" in col or "Salary" in col]
print("Compensation columns:", comp_cols)
```

The line comp_cols = [col for col in df.columns if "Comp" in col or "Salary" in col] automatically collects column names that contain the words “Comp” or “Salary”, so they can be used for further analysis.
